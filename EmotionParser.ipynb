```python
def get_data_column_list(json):
    framedatalist = get_frame_data(json)
    ## 맨 처음 0번째 1143 frame의 데이터프레임
    framedataDF0 = pd.DataFrame(framedatalist[0])
    framedataDF0= framedataDF0.transpose()
    
    Total_Count_Of_Frame = len(framedatalist)
    
    for i in range(1,Total_Count_Of_Frame):
        framedatanext = pd.DataFrame(framedatalist[i])
        framedatanext = framedatanext.transpose()
        framedataDF0 = framedataDF0.append(framedatanext)
    
    return framedataDF0
```

```python
def get_frame_data(json):
    df2 = pd.DataFrame(json)
    df2 = df2[['data']]

    #filteredDF2
    #print('결측치 제거 전 전체 크기 : ',len(filteredDF2))
    df2 = df2.dropna(axis=0)
    #filteredDF2
    #print('결측치 제거 후 전체 크기 : ',len(filteredDF2))
    df2dict = df2.to_dict()
    df2dict_dict = list(df2dict.values())
    
    
    framedatalist = list(df2dict_dict[0].values())
    return framedatalist
```

```python
def get_emotion(json):
    TotalDataFrame = get_data_column_list(json)
    
    PersonEmotionDF = TotalDataFrame.dropna(subset=['emotion'])
    PersonEmotionDF = PersonEmotionDF.drop(['label'],axis=1)
    return PersonEmotionDF



def emotion_parse(json):
    emotion = get_emotion(json)
    length = len(emotion)
    li = []
    for i in range(1,length):
        likeys = list(emotion['emotion'][i].keys())
        for i in range(1,len(likeys)+1):
            likey = likeys[i-1]
            li.append(likey)
    li = set(li)
    li = list(li)
    emotionlist = []
    for i in range(1,length+1):
        for a in li:
            try:
                aka = list(emotion['emotion'][i-1][a].keys())
                for c in range(1,len(aka)+1):
                    ak = aka[c-1]
                    emotionlist.append(ak)
            except:
                continue
    emotionlist = set(emotionlist)
    emotionlist = list(emotionlist)
    
    EmotionListFinal = []
    for i in range(1,length+1):
        for a in li:
            try:
                for b in emotionlist:
                    try:
                        abc = emotion['emotion'][i-1][a]['emotion']
                        EmotionListFinal.append(abc)
                    except:
                        continue
            except:
                continue
    EmotionListFinal = set(EmotionListFinal)
    EmotionListFinal = list(EmotionListFinal)

    return EmotionListFinal
```
